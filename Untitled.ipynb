```python
matrix = []
a_Matrix = [] # alcohol
c_Matrix = [] # control
channels = pd.unique(df["sensor position"])
length = len(files)

for indx in range(length):
    df = pd.read_csv(files[indx])
    sensorData = []
    for ch in channels:
        temp = df.loc[df["sensor position"] == ch]
        #print(len(temp["sensor value"].tolist()))
        sensorData.append(temp["sensor value"].tolist())
        #matrix.append(temp["sensor value"].tolist())
        #print(ch)
    if df.iloc[0, 5] == 'c':
        c_Matrix.append(sensorData)
    else: a_Matrix.append(sensorData)
    matrix.append(sensorData)
    #matrix.append(df["sensor value"].tolist())
    #print(df["sensor value"].tolist())
    
print(len(c_Matrix),len(a_Matrix),len(matrix) ,sep= '  :  ')
```
